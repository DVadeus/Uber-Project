```python
import pandas as pd
```

```python
df = pd.read_csv("uber_data.csv")
```

```python
df.head()
```

```python
df.info()
```

### Convertir object a Datetime

Debido a que ``tpep_pickup_datetime`` y ``tpep_dropoff_datetime`` son tipo de datos object, es necesario convertirlos a tipo de dato **Datetime**

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
```

```python
df.info()
```

### Eliminando duplicados

Eliminaremos los duplicados y posteriormente se resetean los indices ya que al solo eliminar los duplicados los indices no son actualizados.

```python
df = df.drop_duplicates().reset_index(drop=True)
```

Luego insertaremos una nueva columna al incio del dataframe con los valores de los indices, esta columna será usada para crear las tablas de dimensión para el análisis

```python
df.insert(loc=0, column='trip_id', value = df.index)
```

```python
df
```

### Tabla Datetime_dim

Crearemos la tabla dimensional ``Datetime_dim`` en base a las dos columnas ``tpep_pickup_datetime`` y ``tpep_dropoff_datetime``

```python
datetime_dim = df['tpep_pickup_datetime'].to_frame()
datetime_dim['pick_year'] = datetime_dim['tpep_pickup_datetime'].dt.year
datetime_dim['pick_month'] = datetime_dim['tpep_pickup_datetime'].dt.month
datetime_dim['pick_day'] = datetime_dim['tpep_pickup_datetime'].dt.day
datetime_dim['pick_hour'] = datetime_dim['tpep_pickup_datetime'].dt.hour
datetime_dim['pick_weekday'] = datetime_dim['tpep_pickup_datetime'].dt.weekday
datetime_dim['pick_weekday'] = datetime_dim['tpep_pickup_datetime'].dt.weekday
datetime_dim['tpep_dropoff_datetime'] = df['tpep_dropoff_datetime']
datetime_dim['drop_year'] = datetime_dim['tpep_dropoff_datetime'].dt.year
datetime_dim['drop_month'] = datetime_dim['tpep_dropoff_datetime'].dt.month
datetime_dim['drop_day'] = datetime_dim['tpep_dropoff_datetime'].dt.day
datetime_dim['drop_hour'] = datetime_dim['tpep_dropoff_datetime'].dt.hour
datetime_dim['drop_weekday'] = datetime_dim['tpep_dropoff_datetime'].dt.weekday
datetime_dim['drop_weekday'] = datetime_dim['tpep_dropoff_datetime'].dt.weekday
```

```python
datetime_dim
```

Ahora insertaremos la columna PK a la tabla

```python
datetime_dim.insert(loc=0, column='datetime_id', value = datetime_dim.index)
```

```python
datetime_dim
```

### Tabla passenger_count_dim

Tabla dimensional basada en ``passenger_count``

```python
passenger_count_dim = df['passenger_count'].to_frame()
passenger_count_dim.insert(loc=0, column='passenger_count_id', value = passenger_count_dim.index)
```

```python
passenger_count_dim
```

### Tabla trip_distance_dim

Tabla dimensional basada en ``trip_distance``

```python
trip_distance_dim = df['trip_distance'].to_frame()
trip_distance_dim.insert(loc=0, column='trip_distance_id', value = trip_distance_dim.index)
```

```python
trip_distance_dim
```

### Tabla Ratecode_dim

Tabla dimensional basada en ``RatecodeID``, en esta tabla eliminamos los duplicados debido a que solo debe contener los datos diccionario

```python
ratecode_dim = df['RatecodeID'].to_frame().drop_duplicates().reset_index(drop=True)

rate_code_dic = {
    1:"Standard rate",
    2:"JFK",
    3:"Newark",
    4:"Nassau or Westchester",
    5:"Negotiated fare",
    6:"Group ride"
}

ratecode_dim['rate_code_name'] = ratecode_dim['RatecodeID'].map(rate_code_dic)
```

```python
ratecode_dim
```

### Tabla pickup_dim

Tabla dimensional basada en ``pickup_longitude`` y ``pickup_latitude``

```python
pickup_dim = df[['pickup_longitude','pickup_latitude']]
pickup_dim.insert(loc=0, column='pickup_id', value=pickup_dim.index)
```

```python
pickup_dim
```

### Tabla dropoff_dim

Tabla dimensional basada en ``dropoff_longitude`` y ``dropoff_latitude``

```python
dropoff_dim = df[['dropoff_longitude','dropoff_latitude']]
dropoff_dim.insert(loc=0, column='dropoff_id', value=dropoff_dim.index)
```

```python
dropoff_dim
```

### Tabla payment_type_dim

Tabla dimensional basada en ``payment_type``, en esta tabla eliminamos los duplicados debido a que solo debe contener los datos diccionario

```python
payment_type_dim = df['payment_type'].to_frame().drop_duplicates().reset_index(drop=True)

payment_type_dic = {
    1:"Credit card",
    2:"Cash",
    3:"No charge",
    4:"Dispute",
    5:"Unknown",
    6:"Voided trip"
}

payment_type_dim['payment_type_name'] = payment_type_dim['payment_type'].map(payment_type_dic)
```

```python
payment_type_dim.rename(columns={"payment_type_id": "payment_type"}, inplace=True)

payment_type_dim
```

```python
fact_table = df[['trip_id','VendorID']]
```

```python
fact_table
```

```python
fact_table = fact_table \
.join(datetime_dim['datetime_id']) \
.join(passenger_count_dim['passenger_count_id']) \
.join(trip_distance_dim['trip_distance_id']) \
.join(pickup_dim['pickup_id']) \
.join(dropoff_dim['dropoff_id'])
```

```python
fact_table = fact_table.join(df['payment_type'])
```

```python
fact_table
```

```python
fact_table = fact_table.merge(payment_type_dim, how='left').drop(labels=['payment_type'], axis=1)
```

```python
fact_table
```

```python
fact_table = fact_table.join(df['RatecodeID'])
```

```python
fact_table
```

```python
fact_table = fact_table.merge(ratecode_dim, how='left').drop(labels=['RatecodeID'], axis=1)
```

```python
fact_table
```

```python
fact_table = fact_table.join(df['store_and_fwd_flag'])
```

```python
fact_table = fact_table.join(df['extra']).join(df['mta_tax']).join(df['tip_amount']).join(df['tolls_amount']).join(df['improvement_surcharge']).join(df['total_amount'])
```

```python
fact_table_test = fact_table_test \
.merge(datetime_dim, left_on='trip_id', right_on='datetime_id', how='outer') \
.merge(passenger_count_dim, left_on='trip_id', right_on='passenger_count_id', how='outer') \
.merge(trip_distance_dim, left_on='trip_id', right_on='trip_distance_id', how='outer') \
.merge(pickup_dim, left_on='trip_id', right_on='pickup_id', how='outer') \
.merge(dropoff_dim, left_on='trip_id', right_on='dropoff_id', how='outer')
```
